```python
import asyncpg


async def main():
    conn = await asyncpg.connect(
        host="localhost",
        database="bootcamp",
        user="postgres",
        password="postgres",
        port=5400
    )

    try:
        version = await conn.fetchval("SELECT version()")
        print("Versión:", version)

        sales = await conn.fetch("SELECT * FROM sales")
        for sale in sales:
            print(sale['product'], sale['price'])

    finally:
        await conn.close()

await main()
```
